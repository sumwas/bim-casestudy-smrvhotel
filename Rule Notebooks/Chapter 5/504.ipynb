```python
import rdflib
import pandas as pd
# Load RDF graph
rdf_graph = rdflib.Graph()
rdf_graph.parse("..\..\LBD files\smartReview\outputHotel Compliant Revit 2023 Advanced with Spaces.ifc.ttl", format="ttl")
```

```python
# Getting Construction Type, Occupancy Group, & Sprinkler Type for the building
with open("..\..\SPARQL\storeyRule.sparql", "r") as file:
    sparql_query = file.read()

results = rdf_graph.query(sparql_query)

for result in results:
    construction_type = result.constructionType.toPython()
    occupancy_group = result.occupancyGroup.toPython()
    sprinkler_type = result.sprinklerType.toPython() if result.sprinklerType else "none"
    storey_num = result.storeyNumValue.toPython()

    # Splitting the construction type and subtype 
    if "-" in construction_type:
        main_construction_type, subtype = construction_type.rsplit("-", 1)
    else:
        main_construction_type = construction_type
        subtype = ""
if sprinkler_type != "none":
    sprinkler_system = "S" 
else: 
    sprinkler_system = "NS"
       
#print(main_construction_type, subtype, occupancy_group, sprinkler_system)
```

```python
def findVal(table, occupancy_group, sprinkler_system, main_construction_type, subtype):
    height_limit = None  # Initialize height_limit to None before the loop

    # Iterate over all possible keys in the table
    for occupancy in table:
        # Split the key and check if occupancy_group is in the resulting list
        if occupancy_group in [o.strip() for o in occupancy.split(",")]:
            try:
                height_limit = table[occupancy][sprinkler_system][main_construction_type][subtype]
            except KeyError as e:
                print(f"KeyError: {e}")
                height_limit = None
            break

    return height_limit


#print(findVal(table_504_3, occupancy_group, sprinkler_system, main_construction_type, subtype))
```

```python
# Function to check Value
def checkVal(val, allowableVal):
    if allowableVal is None:
        return "Value limit not found"
    
    val = int(val)
    allowableVal = int(allowableVal)
    
    if val <= allowableVal:
        return "Passed"
    else:
        return "Failed"
```

## Individual Occupancy Height Analysis
Each occupancy is checked to determine if the height of the occupancy exceeds the limits defined for the occupancy in Table 504.3. The results are presented in the following table:

```python
with open("..\..\SPARQL\\areasSpaces.txt", "r") as file:
    sparql_query = file.read()

results = rdf_graph.query(sparql_query)

# Prepare data for the table
data = {
    "Story": [],
    "Revit ID": [],
    "Occupancy": [],
    "Name": [],
    "Height": [],
    "Allowable Height": [],
    "Height Result": []

}

# Process SPARQL results
for row in results:
    story = row["levelValue"]
    revit_id = row["id"]
    occupancy = str(row["occupancyClassification"])
    name = row["name"]
    height = float(row["storeyHeightValue"])  # Convert height to float
    allowable_height = findVal(table_504_3, occupancy, sprinkler_system, main_construction_type, subtype)
    height_result = checkVal(height, allowable_height)

    data["Story"].append(story)
    data["Revit ID"].append(revit_id)
    data["Occupancy"].append(occupancy)
    data["Name"].append(name)
    data["Height"].append(f"{height:.2f} ft")
    data["Allowable Height"].append(allowable_height)
    data["Height Result"].append(height_result)
    
    
df = pd.DataFrame(data)
df
```

## Individual Occupancy Story Analysis
Each occupancy is checked to determine if the story location of the occupancy exceeds the limits defined for the occupancy in Table 504.4. The results are presented in the following table:

```python
with open("..\..\SPARQL\\areasSpaces.txt", "r") as file:
    sparql_query = file.read()

results = rdf_graph.query(sparql_query)

# Prepare data for the table
data = {
    "Story": [],
    "Revit ID": [],
    "Occupancy": [],
    "Name": [],
    "Allowable Story": [],
    "Story Result": []

}

# Process SPARQL results
for row in results:
    story = str(row["levelValue"]).lstrip('0')  # Remove leading zeros
    revit_id = row["id"]
    occupancy = str(row["occupancyClassification"])
    name = row["name"]
    allowable_story = findVal(table_504_4, occupancy, sprinkler_system, main_construction_type, subtype)
    story_result = checkVal(story, allowable_story)

    data["Story"].append(story)
    data["Revit ID"].append(revit_id)
    data["Occupancy"].append(occupancy)
    data["Name"].append(name)
    data["Allowable Story"].append(allowable_story)
    data["Story Result"].append(story_result)
    
    
df = pd.DataFrame(data)
df
```

# Section 504.3 - Height in feet
## Building Code
The maximum height, in feet, of a building shall not exceed the limits specified in Table 504.3. 

**Exception**: Towers, spires, steeples and other roof structures shall be constructed of materials consistent with the required type of construction of the building except where other construction is permitted by Section 1510.2.5.
